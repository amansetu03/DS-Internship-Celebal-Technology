```python
import pandas as pd
```

```python
df = pd.read_csv("data/gym.csv")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
features = df.drop(columns=['number_people', 'date', 'timestamp'])

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
#pca 
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

```

```python
# Visualization and Analysis
import matplotlib.pyplot as plt

# Plotting the PCA results
plt.figure(figsize=(10, 3))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=df['number_people'], cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Gym Dataset')
plt.colorbar(label='Number of People')
plt.show()

# Explained variance
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by Principal Component 1: {explained_variance[0]:.2f}")
print(f"Explained variance by Principal Component 2: {explained_variance[1]:.2f}")
```
