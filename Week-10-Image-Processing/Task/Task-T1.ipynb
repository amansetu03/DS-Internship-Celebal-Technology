```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load image
image = cv2.imread('Image/test.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(image)
```

```python
# Increase brightness (add 100 to all pixel values)
brightened_image = cv2.add(image, 100)
plt.axis("off")
plt.imshow(brightened_image)
```

```python
noisy_image = image + np.random.normal(0, 25, image.shape).astype(np.uint8)
plt.axis("off")
plt.imshow(noisy_image)
```

```python
alpha = 1.5  # Contrast factor
contrast_image = cv2.convertScaleAbs(image, alpha=alpha, beta=0)
plt.axis("off")
plt.imshow(contrast_image)
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

plt.axis("off")
plt.imshow(blurred_image)
```

```python
blurred = cv2.GaussianBlur(image, (5, 5), 0)
sharpened_image = cv2.addWeighted(image, 1.5, blurred, -0.5, 0)

plt.axis("off")
plt.imshow(sharpened_image)
```

```python
edges = cv2.Canny(image, 100, 200)

plt.axis("off")
plt.imshow(edges)
```

```python
scale_factor = 2  # Change as needed
resized_image = cv2.resize(image, (image.shape[1] * scale_factor, image.shape[0] * scale_factor))

plt.axis("off")
plt.imshow(resized_image)
```

```python
overlay_image = cv2.imread('Image/test.jpeg')
composite_image = cv2.addWeighted(image, 0.7, overlay_image, 0.3, 0)

plt.axis("off")
plt.imshow(composite_image)
```
